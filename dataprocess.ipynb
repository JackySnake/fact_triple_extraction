```python
import pandas as pd
```

```python
data = pd.read_excel('data/all_illness_symptom.xlsx')
data
```

```python
is_duplicated = data['疾病'].duplicated()
illness = data['疾病'][~is_duplicated]
```

```python
illness
```

```python
is_duplicated = data['症状'].duplicated()
symptom = data['症状'][~is_duplicated]
symptom
```

```python
dic = pd.concat((illness, symptom))
```

```python
aa = dic.tolist()
```

```python
fw = open('dic', 'w')
```

```python
for i in aa:
    fw.write(i + '\n')
fw.close()
```
